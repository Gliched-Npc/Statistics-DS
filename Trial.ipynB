```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt  
from  sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA

data = sns.load_dataset('iris')
data.head(10)

x= data.drop('species',axis=1)
y=data['species'] 

scalar=StandardScaler()
x_sacaled=scalar.fit_transform(x)

pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_sacaled)

pca_data=pd.DataFrame(data=x_pca, columns=['PCa1','PCa2'])
pca_data=pd.concat([pca_data,y.reset_index(drop=True)],axis=1)

print(pca_data)

plt.figure(figsize=(8,5))
plt.scatter(pca_data['PCa1'],pca_data['PCa2'],cmap="viridis")
plt.title('PCA on Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Species')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define a function to plot a 3D hyperplane in R^3 (a 2D plane in 3D space)
def plot_hyperplane_3d():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Create a grid of points for the hyperplane
    x = np.linspace(-10, 10, 50)
    y = np.linspace(-10, 10, 50)
    X, Y = np.meshgrid(x, y)
    
    # Define a hyperplane: ax + by + cz + d = 0
    # For example: x + y + z = 0 -> z = -x - y
    Z = -X - Y
    
    # Plot the hyperplane
    ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='cyan', edgecolor='k')
    
    # Labels and limits
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('A Hyperplane in 3D Space (Plane in R^3)')
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_zlim(-10, 10)
    
    # Show the plot
    plt.show()

plot_hyperplane_3d()
```
